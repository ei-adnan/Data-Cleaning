```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_sales.csv')
```

```python
# Top 10 items by revenue

top_items = df.groupby('Item')['Total Spent'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_items.index, top_items.values)
plt.title("Top 10 Items by Revenue")
plt.xlabel("Item")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Payment method distribution

payment_counts = df['Payment Method'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(payment_counts.values, labels=payment_counts.index, autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()
```

```python
# Revenue by payment method

payment_revenue = df.groupby('Payment Method')['Total Spent'].sum()

plt.figure(figsize=(8,5))
plt.bar(payment_revenue.index, payment_revenue.values)
plt.title("Revenue by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Price vs Quantity scatter plot

sample = df.sample(500)

plt.figure(figsize=(8,5))
plt.scatter(sample['Price Per Unit'], sample['Quantity'])
plt.title("Price vs Quantity")
plt.xlabel("Price Per Unit")
plt.ylabel("Quantity")
plt.tight_layout()
plt.show()
```
