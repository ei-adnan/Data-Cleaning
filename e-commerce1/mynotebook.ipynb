```python
"""After removing the duplicates the index 
is 27931 but we still have 30757 rows, lets fix that."""
```

```python
data.tail()
```

```python
data = data.reset_index()
```

```python
data.head() 
```

```python
"""Now we have an Index column that is unnecessary, we can drop it."""

data.drop(columns=['index'], inplace=True)
```

```python
data
```

```python
"""Lets Capitalize the Brand names in the BrandName column"""

data['BrandName'] = data['BrandName'].apply(lambda x: x.capitalize())
```

```python
data['Discount'].value_counts(dropna=False)
```

```python
"""We have thousands of rows in the Discount column filled as Nan,
 which are not showing up in the value counts,
Lets fix it. We can fill it with the median value 
of the Discount column but grouped by the category of the product, 
so that we can have a more accurate representation of the discounts for each category."""
```

```python
"""Replacing all the rows that have nan with the actual NaN value from numpy, 
so that we can use the fillna method to fill them with the median value."""

import numpy as np

data['Discount'] = data['Discount'].replace('Nan', np.nan)
```

```python
data
```

```python
"""Right now the Discount column is still of object type, 
we need to convert it to numeric type before filling the NaN values with the median.
Also, we need to extract the numeric part of the discount values, 
since they are currently in the format of '20%'. 
We can use the str.extract method."""

data['Discount'] = data['Discount'].str.extract(r'(\d+)').astype(float)
```

```python
data
```

```python
data['Discount'] = data['Discount'].fillna(data.groupby('Category')['Discount'].transform('median'))
```

```python
data['Discount'].isna().sum()

"""Now we have successfully filled all the NaN values in the Discount column with 
the median value of the discounts for each category."""
```

```python
data['Discount'].value_counts(dropna=False)
```

```python
data
```

```python
"""Now we do the same for the MRP column, we have some Nan 
values that we need to fill with the median value of the MRP for each category."""
```

```python
data['MRP'] = data['MRP'].replace('Nan', np.nan)
```

```python
data['MRP'] = data['MRP'].astype(str).str.replace(r'Rs\s*', '', regex=True)
data['MRP'] = data['MRP'].str.replace(r'\n', '', regex=True)
```

```python
data['MRP'].head()
```

```python
data.head(40)
```

```python
import numpy as np

data['MRP'] = data['MRP'].replace(
    ['nan', 'NaN', 'Nan', 'NULL', 'null', 'None', ''],
    np.nan
)
```

```python
data['MRP'].isna().sum()
```

```python
"""We have successfully cleaned the MRP column by removing the 'Rs' prefix and any newline characters,
and replacing any non-standard representations of missing values with actual NaN values.
Now we can fill the NaN values in the MRP column with the median value of the MRP for each category, 
just like we did with the Discount column."""
```

```python
data['MRP'] = data['MRP'].fillna(data.groupby('Category')['MRP'].transform('median'))
```

```python
data['MRP'].isna().sum()
```

```python
data
```

```python
data.to_csv('Cleaned_FashionDataset.csv', index=False)
```
